```python
question7_df = data.loc[data.ocean_proximity == "ISLAND",['housing_median_age','total_rooms','total_bedrooms']]
X = question7_df.to_numpy()
XTX = X.T.dot(X)
inv_XTX = np.linalg.inv(XTX)
y = np.array([950, 1300, 800, 1000, 1300])
w = (inv_XTX.dot(X.T)).dot(y)
w
```
